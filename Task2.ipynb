```python
# Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#set numbers for how many rows and columns will be displayed
pd.set_option('display.min_rows', 10) #default will be 10
pd.set_option('display.max_columns', 20)


df = pd.read_csv('/Au_nanoparticle_dataset.csv')


new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]



# Display the first 20 samples
print(new_df.head(20))
```

```python
# Calculate mean, standard deviation, and quartiles
statistics = new_df.describe()

    # Display the calculated statistics
print("Mean values:")
print(statistics.loc['mean'])
print("\nStandard deviation values:")
print(statistics.loc['std'])
print("\nQuartiles:")
print(statistics.loc[['25%', '50%', '75%']])

```

```python
# Plot histograms in a 1x4 layout
plt.figure(figsize=(20, 5))

for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    sns.histplot(new_df[column], kde=True)
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

```python
# Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()
```

```python
# Create a PairGrid with customized plots
g = sns.PairGrid(new_df)
```

```python
# Diagonal plots: Histograms with kernel density estimation
g.map_diag(sns.histplot, kde=True)
```

```python
# Add titles to diagonal plots
for i in range(len(new_df.columns)):
    g.axes[i, i].set_title(new_df.columns[i])

plt.tight_layout()
plt.show()
```

```python
# Create a PairGrid with customized plots
g = sns.PairGrid(new_df)

# Diagonal plots: Histograms with kernel density estimation
g.map_diag(sns.histplot, kde=True)

# Lower triangle plots: Bivariate kernel density estimation
g.map_lower(sns.kdeplot)

# Add titles to diagonal plots
for i in range(len(new_df.columns)):
    g.axes[i, i].set_title(new_df.columns[i])

plt.tight_layout()
plt.show()
```
